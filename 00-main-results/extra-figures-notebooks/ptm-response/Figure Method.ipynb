```python
def custom_order_predictor_edges(predictor):
    
    overrides = {
        'H3K4me3': [['H02'], ['H15', 'H11'], ['H13', 'H21'], ['H14', 'H22'], ['H40', 'H41']],
        'H3K4me1': [['H05'], ['H15', 'H09'], ['H13', 'H16'], ['H30', 'H31'], ['H03', 'H06']],
    }
    
    selected = predictor_edges(predictor)
    
    if predictor not in overrides:
        return selected
    else:
        # Check if we did not break anything
        set_selected = set(['-'.join(s) for s in selected])
        overriden = overrides[predictor]
        
        set_overriden = set(['-'.join(s) for s in overriden])
        
        assert set_selected == set_overriden
        
        return overriden
    
```

```python
def edge_tostring(edge):
    if len(edge) == 1:
        return '{}-(self)'.format(edge[0])
    else:
        return '-'.join([edge[1], edge[0]])
    
def plot_protein_response_bars(ptm, 
                               proteins_to_highlight,
                               width=0.984252, # 25mm
                               height=0.3937, # 10mm,
                               ylims=None):
    
    lm = load_limma_long_matrix(ptm)
    limma_result = load_limma_data(ptm)

    protein_order = list(proteins_to_highlight.keys())

    edges = custom_order_predictor_edges(ptm)
    edges = list(map(edge_tostring, edges))

    n_proteins = len(protein_order)
    
    labelsize = 5
    
    with sns.plotting_context({
        'axes.linewidth': 0.5,
        'grid.linewidth': 0.1,
        'grid.linestyle': ':',
        'xtick.labelsize': labelsize,
        'ytick.labelsize': labelsize,
        'xtick.major.size': 2,
        'xtick.major.width': 0.5,
        'xtick.major.pad': 2,
        'xtick.minor.size': 1.5,
        'xtick.minor.width': 0.2,
        'xtick.minor.pad': 2,
        
        'ytick.major.size': 2,
        'ytick.major.width': 0.5,
        'ytick.major.pad': 2,
        'ytick.minor.size': 1.5,
        'ytick.minor.width': 0.2,
        'ytick.minor.pad': 2,
        
        'axes.labelsize': labelsize,
        'axes.titlesize': labelsize,
        
    }):
    
        fig = plt.figure(figsize=(width, n_proteins * height), 
                         dpi=300,
                         constrained_layout=True)

        gs = gridspec.GridSpec(ncols=1, nrows=n_proteins, figure=fig)

        if not isinstance(ylims, list):
            ylims = [ylims] * n_proteins

        ax = None

        for plot_row, (protein, ylim) in enumerate(zip(protein_order, ylims)):
            color = proteins_to_highlight[protein]

            ax = fig.add_subplot(gs[plot_row, 0])

            subdf = lm.loc[protein, edges, True, :].reset_index()
            

            sns.barplot(x='edge', y='normed_ratio', 
                        order=edges, 
                        color=color, 
                        ax=ax,
                        errwidth=0.5,
                        capsize=0.25,
                        ci=None,
                        data=subdf.reset_index())
            
            

            sns.swarmplot(x='edge', y='normed_ratio', 
                        order=edges, 
                        color='black', 
                        ax=ax,
                        size=1.0,
                        marker='o',
                        data=subdf.reset_index())

        #     ax.axhline(0, linestyle='-', color='k ', linewidth=2, zorder=3)

            if ylim is None:
                ylim = list(ax.get_ylim())

                ylim[0] = np.sign(ylim[0]) * np.ceil(np.abs(ylim[0]) * 2) / 2 
                ylim[1] = np.sign(ylim[1]) * np.ceil(np.abs(ylim[1]) * 2) / 2 

            absmax = np.max(np.abs(ylim))

            ax.set_ylim(ylim)

            if absmax < 1:  
                ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))
            else:
                ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
                ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.5)) 

            sns.despine(ax=ax, bottom=True, offset=5)
            ax.set_xlabel('')
            ax.set_ylabel('')
            ax.set_title('')
            

#             ax.set_title(protein)
            ax.yaxis.set_label_position("right")
            ax.set_ylabel(protein, rotation=0, ha='left', va='center', 
                          fontsize=labelsize+1, fontweight="bold")
            ax.xaxis.set_tick_params(length=0, rotation=90)

            ax.axhline(limma_result.loc[protein, 'logFC'], color='black', linewidth=1.0)


            if plot_row != n_proteins - 1:
                for tick in ax.get_xticklabels():
                    ax.xaxis.set_tick_params(length=0)
                    tick.set_visible(False)
        
        fig.set_constrained_layout_pads(w_pad=0.00, h_pad=0.00, wspace=0.05, hspace=0.05)
        
        return fig
    
```

```python
def plot_scatter_travels_horizontal(enrichment_data, predictor, proteins_to_highlight, 
                                    xlim=None, ylim=None, annot_axes=True, 
                                    sizing='tiny'):
    
    data = enrichment_data.copy()
    data['group'] = 'other'

    for prot in proteins_to_highlight:
        data.loc[prot, 'group'] = prot

    marker_imputed = '.'
    marker_standard = 'o'
    
    

    
    color_other = '#d5d5d5'
    alpha_other = .5
 
    edges = custom_order_predictor_edges(predictor)
    n_edges = len(edges)
    
    
    if sizing == 'tiny':
    
        marker_size_imputed = 5
        marker_size_standard = 10
        width = height = 0.55    
        labelsize = 5
        edgewidth = 0.5
        
        divider_line_width = 1.0
        
        context = {
            'axes.linewidth': 0.5,
            'grid.linewidth': 0.1,
            'grid.linestyle': ':',
            'xtick.labelsize': labelsize,
            'ytick.labelsize': labelsize,
            'xtick.major.size': 2,
            'xtick.major.width': 0.5,
            'xtick.major.pad': 2,
            'xtick.minor.size': 1.5,
            'xtick.minor.width': 0.2,
            'xtick.minor.pad': 2,

            'ytick.major.size': 2,
            'ytick.major.width': 0.5,
            'ytick.major.pad': 2,
            'ytick.minor.size': 1.5,
            'ytick.minor.width': 0.2,
            'ytick.minor.pad': 2,
        }
    elif sizing == 'huge':
        marker_size_imputed = 5*6
        marker_size_standard = 10*6
        width = height = 1.5748 # 40mm
        labelsize = 5
        edgewidth = 1.5
        
        divider_line_width = 2.0
        
        context = {
            'axes.linewidth': 1.0,
            'grid.linewidth': 0.2,
            'grid.linestyle': ':',
            'xtick.labelsize': labelsize,
            'ytick.labelsize': labelsize,
            'xtick.major.size': 4,
            'xtick.major.width': 0.6,
            'xtick.major.pad': 3,
            'xtick.minor.size': 3,
            'xtick.minor.width': 0.4,
            'xtick.minor.pad': 3,

            'ytick.major.size': 4,
            'ytick.major.width': 0.6,
            'ytick.major.pad': 3,
            'ytick.minor.size': 3,
            'ytick.minor.width': 0.4,
            'ytick.minor.pad': 3,
        }
    else:
        raise ValueError("Supported sizes are tiny or huge, got {sizing!r}")
    
    with sns.plotting_context(context):

        fig = plt.figure(figsize=(width*n_edges, 2 * height), 
                         dpi=300,
                         constrained_layout=True)

        gs = gridspec.GridSpec(ncols=n_edges, nrows=2, figure=fig)

        ax_left = None
        ax_right = None

        for plot_col, edge in enumerate(edges):

            if len(edge) == 1:
                ax_left = None
                ax_right = fig.add_subplot(gs[1, plot_col], 
                                    sharex=ax_right, 
                                    sharey=ax_right)
                subdata_row = data.loc(axis=0)[:, edge]
                pd_left = None
                pd_right = edge[0]

            else:
                ax_left = fig.add_subplot(gs[0, plot_col], 
                                    sharex=ax_right, 
                                    sharey=ax_right)
                ax_right = fig.add_subplot(gs[1, plot_col], 
                                    sharex=ax_right, 
                                    sharey=ax_right)

                subdata_row = data.loc(axis=0)[:, edge]
                pd_left, pd_right = edge

            # First plot "others"
            subdata_others = subdata_row[subdata_row['group'] == 'other']
            for (pd, is_imputed), sdf in subdata_others.reset_index().groupby(('Pull-Down ID', 'is_imputed')):
                ax = ax_left if pd == pd_left else ax_right
                ax.scatter(sdf[COLUMN_X], 
                           sdf[COLUMN_Y],
                           s=marker_size_imputed,
                           alpha=alpha_other,
                           color=color_other,
                           marker=marker_imputed if is_imputed else marker_standard,
                           rasterized=True)

            # Now let's go through the remaining proteins
            for protein, color in proteins_to_highlight.items():

                df_protein = subdata_row.loc[protein]


                if pd_left is not None:
                    row_left = df_protein.loc[pd_left]
                else:
                    row_left = None

                row_right = df_protein.loc[pd_right]

                for ax, row in [(ax_left, row_left), (ax_right, row_right)]:
                    if row is not None:
                        ax.scatter(row[COLUMN_X], 
                                    row[COLUMN_Y],
                                    s=marker_size_standard,
                                    alpha=1.0,
                                    facecolor=color,
                                    edgecolor='black',
                                    linewidth=edgewidth / 2,
                                    marker=marker_imputed if row['is_imputed'] else marker_standard,
                                    rasterized=True)


                if row_left is not None:
                    prev_x, prev_y, prev_imputed = row_left[COLUMN_X], row_left[COLUMN_Y], row_left['is_imputed']
                else:
                    prev_x = prev_y = 0
                    prev_imputed = False


                ax_right.scatter(prev_x, 
                                 prev_y,
                                 s=marker_size_standard,
                                 alpha=1.0,
                                 facecolor='none',
                                 edgecolor=color,
                                 linewidth=edgewidth,
                                 marker=marker_imputed if prev_imputed else marker_standard,
                                 rasterized=True)


                dx = row_right[COLUMN_X] - prev_x
                dy = row_right[COLUMN_Y] - prev_y

                if annot_axes:
                    if ax_left is not None:
                        ax_left.text(0.97, 0.97, pd_left, fontweight='bold', 
                                     fontsize=labelsize,
                                     transform=ax_left.transAxes, ha='right', va='top')

                    ax_right.text(0.97, 0.97, pd_right, fontweight='bold', 
                                  fontsize=labelsize,
                                  transform=ax_right.transAxes, ha='right', va='top')

            for ax in [ax_left, ax_right]:
                if ax is None:
                    continue
                for spine in ax.spines.values():
                    spine.set_visible(True)

                ax.axvline(0, color='black', zorder=0, linewidth=divider_line_width)
                ax.axhline(0, color='black', zorder=0, linewidth=divider_line_width)

                if xlim is not None:
                    ax.set_xlim(*xlim)
                if ylim is not None:
                    ax.set_ylim(*ylim)

                ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
                ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))
                ax.yaxis.set_major_locator(ticker.MultipleLocator(2))
                ax.yaxis.set_minor_locator(ticker.MultipleLocator(1))
                ax.set_aspect('equal')                   

        fig.set_constrained_layout_pads(w_pad=0.00, h_pad=0.00, wspace=0.03, hspace=0.03)

    return fig
```
